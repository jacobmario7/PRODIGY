```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/drive/MyDrive/data science/age_gender.csv")
data
```

```python
plt.hist(data['age'],color='red',edgecolor='black',bins=8)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Age Distribution")
```

```python
sns.countplot(x='gender',data=data,palette="Set1",edgecolor='black',hue='gender',legend=False)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("Gender Distribution")
```
