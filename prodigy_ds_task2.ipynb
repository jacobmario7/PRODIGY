```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/drive/MyDrive/data science/gender_submission.csv")
data
```

```python
data['Survived'].fillna(data['Survived'].median(), inplace=True)
```

```python
survived=data['Survived'].value_counts()
survived
```

```python
sns.countplot(x='Survived',data=data,palette="Set2",edgecolor='black')
plt.title('Survival Count')
plt.show()
```

```python
sns.barplot(x='PassengerId',y='Survived', data=data)
plt.figure(figsize=(8, 6))
plt.show()
```
